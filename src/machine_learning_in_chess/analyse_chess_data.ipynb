### Distribution of Player Ratings


```python
# Merge the counts into one, as the player's side isn't important.
white_rating_count = df["WhiteElo"].value_counts().compute()
black_rating_count = df["BlackElo"].value_counts().compute()
rating_counts = (
    pd.concat([white_rating_count, black_rating_count], axis=1).fillna(0).astype(int)
)
rating_counts = rating_counts.sum(axis=1)
# Find the minimum and maximum player rating.
max_rating = rating_counts.index.max()
min_rating = rating_counts.index.min()
print(f"Maximum Player Rating: {max_rating}")
print(f"Minimum Player Rating: {min_rating}")

# Define the bin edges for player ratings from 600 (minimum in Lichess) to
# 3600 in intervals of 200.
rating_bins = np.arange(600, 3601, 200)
rating_counts.index = pd.cut(rating_counts.index, bins=rating_bins)
merged_rating_counts = rating_counts.groupby(level=0).sum()
display(merged_rating_counts)
```

```python
# Plot a graph to show the distribution of player ratings.
sns.set(font_scale=1.25)
fig, ax = plt.subplots(figsize=(10, 6))
indexes = merged_rating_counts.index.astype(str)
values = merged_rating_counts.values
ax.bar(indexes, values)
ax.set_title("Distribution of Player Ratings on Lichess in 2022")
ax.set_xlabel("Lichess Player Rating Range")
ax.set_ylabel("Number of Players")
ax.set_xticklabels(indexes, rotation=60)
plt.show()
```

### Rating Difference of Matched Players


#### Create Rating Difference Column


```python
# Create an EloDiff column to represent the difference in Elo between the two
# players.
df["EloDiff"] = df["WhiteElo"] - df["BlackElo"]
# Calculate the relative rating difference of White to Black as a percentage.
df["RelativeEloDiff"] = (
    ((df["WhiteElo"] / (df[["WhiteElo", "BlackElo"]].median(axis=1))) - 1.0) * 100.0
).round(2)
display(df.head())
```

## Prediction of Game Results Using Classifiers


### Decision Tree Classifier


#### Baseline Model  Only RelativeEloDiffBin


```python
# Re-bin the RelativeEloDiff values into more precise bins to improve the
# model's efficacy.
relative_elo_diff_bins = np.arange(-60, 61, 5)
# Create a new column that says what RelativeEloDiff bin each game belongs to.
df_reduced = df_reduced.map_partitions(
    lambda df_reduced: df_reduced.assign(
        RelativeEloDiffBin=lambda df_reduced: pd.cut(
            df_reduced["RelativeEloDiff"], bins=relative_elo_diff_bins
        )
    )
)
display(df_reduced[["RelativeEloDiff", "RelativeEloDiffBin"]].head())

# Select features.
X = df_reduced[["RelativeEloDiffBin"]].compute()
y = df_reduced["Result"].compute()

# Encode categorical features to numerical using one-hot encoding.
X = pd.get_dummies(X)

# Split the data into training and test sets.
X, X_test, y, y_test = train_test_split(X, y, test_size=0.2, random_state=69)

# Create and fit the decision tree classifier.
clf = DecisionTreeClassifier(random_state=69)
clf.fit(X, y)

# Make predictions on the test set.
y_pred = clf.predict(X_test)
# Evaluate the model's performance.
acc = accuracy_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred, average="weighted")
print(f"Accuracy of the decision tree classifier: {acc:.3f}")
print(f"F1 score of the decision tree classifier: {f1:.3f}")
```

## Cluster Analysis of Openings by Game Results


### Optimisation of K-Means Clustering


```python
# Assume result_percentages_pivot_sorted is already available from previous
# steps.
clustering_data = result_percentages_pivot_sorted[["1-0", "1/2-1/2", "0-1"]]
# Convert the Dask DataFrame to a Pandas DataFrame for compatibility with scikit-learn.
clustering_data_pd = pd.DataFrame(clustering_data)

# Determine the range of cluster numbers to test.
min_clusters = 3
max_clusters = 10

# Calculate WCSS (Within-Cluster Sum of Squares) for each number of clusters.
wcss_vals = []
silhouette_score_vals = {}
for i in range(min_clusters, max_clusters + 1):
    kmeans = KMeans(n_clusters=i, init="k-means++", random_state=69)
    kmeans.fit(clustering_data_pd)
    wcss_vals.append(kmeans.inertia_)
    silhouette_score_vals[i] = silhouette_score(clustering_data_pd, kmeans.labels_)

# Plot the WCSS values.
sns.set(font_scale=1.5)
plt.figure(figsize=(10, 6))
plt.plot(range(min_clusters, max_clusters + 1), wcss_vals, marker="o")
plt.xlabel("Number of Clusters")
plt.ylabel("Within-Cluster Sum of Squares (WCSS)")
plt.title(
    "Elbow Method for Optimal Number of Clusters for\nK-Means Clustering of Base Openings by Game Results"
)
plt.xticks(np.arange(min_clusters, max_clusters + 1))
plt.grid()
plt.show()

# Show the table of silhouette scores.
table_of_silhouette_scores = pd.DataFrame.from_dict(
    silhouette_score_vals,
    columns=["Silhouette Score"],
    orient="index",
)
table_of_silhouette_scores.reset_index(inplace=True)
table_of_silhouette_scores.columns = ["Number of Clusters", "Silhouette Score"]
display(table_of_silhouette_scores.style.hide_index())
```

### Calculation of Proportions of Game Results Per Opening Variation


```python
# Count the number of games for each opening.
opening_counts = df_reduced["Opening"].value_counts().compute()
# Filter the openings that have over 5,000 games.
popular_openings = opening_counts[opening_counts >= 5000].index
filtered_df = df_reduced[df_reduced["Opening"].isin(popular_openings)]

# Calculate the total games per opening in the filtered DataFrame.
total_games = filtered_df.groupby("Opening").size().compute()
# Count the games for each result per opening.
result_counts = filtered_df.groupby(["Opening", "Result"]).size().compute()
# Calculate the percentage for each result per opening.
result_percentages = (result_counts / total_games * 100).reset_index(name="Percentage")
# Pivot the result percentages for a clearer view.
result_percentages_pivot_opening = result_percentages.pivot_table(
    index="Opening", columns="Result", values="Percentage"
).reset_index()
# Add a column to show the number of games for each opening.
result_percentages_pivot_opening["NumberOfGames"] = result_percentages_pivot_opening[
    "Opening"
].map(total_games)
# Convert filtered_df to a pandas DataFrame.
filtered_df_pd = filtered_df.compute()

# Merge 'BaseOpening' column from filtered_df_pd based on 'Opening'.
result_percentages_pivot_opening = result_percentages_pivot_opening.merge(
    filtered_df_pd[["Opening", "BaseOpening"]].drop_duplicates(),
    on="Opening",
    how="left",
)
# Sort the table by the number of games descending.
result_percentages_pivot_opening_sorted = result_percentages_pivot_opening.sort_values(
    "NumberOfGames", ascending=False
)
# Delete rows with BaseOpenings that aren't recognized or are too generic.
result_percentages_pivot_opening_sorted = result_percentages_pivot_opening_sorted[
    ~result_percentages_pivot_opening_sorted["BaseOpening"].isin(
        ["?", "King's Pawn Game", "Queen's Pawn Game"]
    )
]
result_percentages_pivot_opening_sorted.reset_index(drop=True, inplace=True)
display(result_percentages_pivot_opening_sorted)
```

### Calculation of Mean Euclidean Distances of Results Between Variations


```python
# Calculate Euclidean distance of results between variations of each
# BaseOpening.
def calculate_mean_euclidean_distance_by_result(group):
    features = group[["1-0", "0-1", "1/2-1/2"]].values
    distances = pdist(features, metric="euclidean")
    return distances.mean()


# Group by BaseOpening and calculate their mean Euclidean distances between
# variations.
mean_distances = result_percentages_pivot_opening_sorted.groupby("BaseOpening").apply(
    calculate_mean_euclidean_distance_by_result
)
# Convert the result to a DataFrame.
mean_distances_df = mean_distances.reset_index(name="MeanEuclideanDistance")

# Sort the DataFrame by MeanEuclideanDistance.
mean_distances_sorted = mean_distances_df.sort_values(
    "MeanEuclideanDistance", ascending=False
)
mean_distances_sorted.dropna(subset=["MeanEuclideanDistance"], inplace=True)
mean_distances_sorted.reset_index(drop=True, inplace=True)
display(mean_distances_sorted)
# Sort in ascending order.
mean_distances_sorted_asc = mean_distances_df.sort_values(
    "MeanEuclideanDistance", ascending=True
)
mean_distances_sorted_asc.dropna(subset=["MeanEuclideanDistance"], inplace=True)
mean_distances_sorted_asc.reset_index(drop=True, inplace=True)
display(mean_distances_sorted_asc)
```
